# Homework on Huber Loss in NumPy

Imports:

```python
import numpy as np
from matplotlib import pyplot as plt, animation, rc
from sklearn.metrics import mean_squared_error
```

Define the regression function. 

```python
def regression_function(alpha,beta,gamma,x): 
    return alpha+beta*x+gamma*x**2
```

Generate artificial data using the "true" values of the regression parameters. The random seed is set to a specific value to ensure reproducibility of the results. Do not modify any parameters so that the homework submission form correctly validates your results. But of couse for your own learning purposes, you are encouraged to modify the parameter values and see the consequences of the changes.

```python
number_of_observations = 50

alpha_true = 3.0
beta_true = -20.0
gamma_true = 32.0
noise_standard_deviation = 0.25

np.random.seed(2)

x_data = np.random.random(number_of_observations)

additional_noise = 8*np.random.binomial(1, 0.03, size=number_of_observations)

y_data = (regression_function(alpha_true,beta_true,gamma_true,x_data)
          + noise_standard_deviation*np.random.randn(number_of_observations) + additional_noise)
```

Plot the generated data. The setup of this figure is later used to create an animation of the optimization process.

```python
fig = plt.figure(dpi=80,figsize=(7,7))
ax = fig.add_subplot(111)
ax.set_xlim((0,1))
ax.set_ylim((-5,20))
plt.scatter(x_data,y_data)
line_thickness = 2
line, = ax.plot([],[],lw=line_thickness)
```

Specify an initial guess for the regression parameter value. Please use gamma_initial_guess = -8.0.

```python
alpha_initial_guess = 3.
beta_initial_guess = -2.
gamma_initial_guess = -8.
```

Choose a learning rate and an animation characteristic.

```python
learning_rate = 0.05
number_of_gradient_descent_steps_per_frame = 50
```

Specify functions that perform a gradient descent step and functions used for creating an animation of the optimization process. You need to modify the gradient_descent_step function to correspond to Huber loss with delta = 2.

```python
delta = 2

def gradient_descent_step():
    global alpha, beta, gamma
    y_predicted = regression_function(alpha,beta,gamma,x_data)
    z = y_data - y_predicted
    
    derivative_of_loss_wrt_alpha = np.where(abs(z)<=delta, (2*(y_predicted-y_data)), (-2*delta*np.sign(z))).mean()
    derivative_of_loss_wrt_beta = np.where(abs(z)<=delta, (2*x_data*(y_predicted-y_data)), (-2*delta*x_data*np.sign(z))).mean()
    derivative_of_loss_wrt_gamma = np.where(abs(z)<=delta, (2*x_data**2*(y_predicted-y_data)), (-2*delta*x_data**2*np.sign(z))).mean()
    
    alpha = alpha - learning_rate*derivative_of_loss_wrt_alpha
    beta = beta - learning_rate*derivative_of_loss_wrt_beta
    gamma = gamma - learning_rate*derivative_of_loss_wrt_gamma


def frame_evaluation(_):
    x_plotting = np.linspace(0,1,51)
    y_plotting = regression_function(alpha,beta,gamma,x_plotting)
    line.set_data(x_plotting,y_plotting)
    for _ in range(number_of_gradient_descent_steps_per_frame): 
        gradient_descent_step()
    return (line,)
```

Set up the animation.

```python
rc('animation',html='html5')
number_of_frames = 250
frame_interval = 100
resulting_animation = animation.FuncAnimation(
    fig,frame_evaluation,np.arange(0,number_of_frames),interval=frame_interval)
```

Set the regression parameter values to the initial guesses. Run the optimization process, capture it as an animation, and display it in the notebook.

```python
alpha = alpha_initial_guess
beta = beta_initial_guess
gamma = gamma_initial_guess

resulting_animation
```

Print the final parameter values.

```python
print('alpha: ',alpha)
print('beta: ',beta)
print('gamma: ',gamma)
```

Save the animation as an mp4 file to the working directory. (Requires resetting the initial values of the parameters.)

```python
alpha = alpha_initial_guess
beta = beta_initial_guess
gamma = gamma_initial_guess

Writer = animation.writers['ffmpeg']
writer = Writer(fps=15, metadata=dict(artist=''), bitrate=1800)
resulting_animation.save('resulting_animation.mp4', writer=writer)
```

```python

y_predicted_initial = regression_function(alpha, beta, gamma, x_data)
initial_mse = mean_squared_error(y_data, y_predicted_initial)
print('Initial MSE without Huber loss: ', initial_mse)
```

```python
y_predicted_huber = regression_function(alpha, beta, gamma, x_data)
huber_mse = mean_squared_error(y_data, y_predicted_huber)
print('MSE with Huber loss: ', huber_mse)
```
