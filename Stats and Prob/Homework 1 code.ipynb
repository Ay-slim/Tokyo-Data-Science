```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
%matplotlib inline
```

```python
def factorial(num):
    """
    Uses recursion to calculate the factorial of a number
    args: number
    returns: factorial
    """
    if num<0:
        return "Factorial doesn't exist"
    elif num == 0:
        return 1
    else:
        return num*factorial(num-1)
```

```python
def binomial_pmf(n, p, k):
    """
    Calculates the Probability mass function of the binomial distribution (estimate of the 
    chances of getting k successes)
    args: n- number of trials
          p- probability of success
          k- number of successes for which we're estimating probability
    returns: pmf- probabilty of k successes
    """
    n_k = factorial(n)/(factorial(k)*factorial(n-k))
    pmf = n_k*(p**k)*((1-p)**(n-k))
    return pmf
```

```python
binomial_pmf(10, 0.3, 4)
```

```python
k_test = [j for j in range(0, 11)]
```

```python
binomial_PMF_value = [binomial_pmf(10,0.3, i) for i in k_test]
```

```python
plt.plot(k_test, binomial_PMF_value)
plt.grid()
```

```python
def poisson_pmf(lamda, k):
    """
    Calculates the probability mass function of the poisson distribution
    args: k- 
    """
    pmf = ((lamda**k)*np.exp(-lamda))/(factorial(k))
    return pmf
```

```python
poisson_pmf(5, 4)
```

```python
poisson_PMF_value = [poisson_pmf(5, i) for i in k_test]
plt.plot(k_test, poisson_PMF_value)
plt.grid()
```
